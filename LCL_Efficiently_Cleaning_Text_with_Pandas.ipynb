```python
HyVee = re.compile(r'(.)*HY-VEE(.)*', 
              re.IGNORECASE)
Costco = re.compile(r'(.)*Costco(.)*', 
              re.IGNORECASE)
Sams = re.compile(r'(.)*Sam’s(.)*', 
              re.IGNORECASE)
```

```python
df.replace({'Store Name': HyVee}, {'Store Name': 'Hy-Vee'}, regex=True, inplace=True)
df.replace({'Store Name': Costco}, {'Store Name': 'Costco'}, regex=True, inplace=True)
df.replace({'Store Name': Sams}, {'Store Name': 'Sam’s'}, regex=True, inplace=True)
```

```python
# create a new column called statut that will contain four categories where Pack column:
# tier_4 — 0-5 
# tier_3 — 6-10 
# tier_2 — 11-20 
# tier_1 — 20+ 
```

```python
df['statut']=pd.cut(df.Pack,[0,5,10,20,df.Pack.max()],labels=['tier_4','tier_3','tier_2','tier_1'])
```

```python
def tiers(pack):
  if pack <=5:
    return 'tier_4'
  elif pack <=10:
    return 'tier_3'
  elif pack <=20:
    return 'tier_2'
  else:
    return 'tier_1'
```

```python
df['stat2'] = df.Pack.apply(tiers)
```

```python
# create a list of our conditions
conditions = [
    (df['Pack'] <= 5),
    (df['Pack'] > 5) & (df['Pack'] <= 10),
    (df['Pack'] > 10) & (df['Pack'] <= 20),
    (df['Pack'] > 20)
    ]

# create a list of the values we want to assign for each condition
values = ['tier_4', 'tier_3', 'tier_2', 'tier_1']

# create a new column and use np.select to assign values to it using our lists as arguments
df['tier'] = np.select(conditions, values)
```
